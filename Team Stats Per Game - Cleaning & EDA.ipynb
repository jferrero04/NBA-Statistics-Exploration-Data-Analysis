```python
# What is the correlation between field goal percentage (FG%) and free throw percentage (FT%) for the Minnesota Timberwolves
# from 2014-2019?
teamOfInterest = ['Minnesota Timberwolves']
seasonsOfInterest = [2014, 2015, 2016, 2017, 2018, 2019,]
FGandFT = ['FG%', 'FT%']

# This line of code, filters the original DF using the variables from above to tailor our new, filtered DF to our exact
# specifications and saves it to the variable 'filteredTeamStatsDF' which can be deployed later.
#
# Two large arguments are joined by the AND bitwise operator (&) to show what we want said team and said seasons (in years)
# to our new DF 'filteredTeamStatsDF'.
timberwolvesFilteredTeamStatsDF = teamStatsDF[(teamStatsDF['Team'].isin(teamOfInterest)) & (teamStatsDF['Season'].isin(seasonsOfInterest))]

# Now we calculate the correlation based on the filtered DF we created.
timberwolvesFGandFTcorrelation = timberwolvesFilteredTeamStatsDF[FGandFT].corr()

display(timberwolvesFGandFTcorrelation)


# This correlation coefficient tells us that the Minnesota Timberwolves had a super high (positive, almost perfect)
# correlation between FG% and FT%. Meaning, as FG% increased, their FT% also increased. Although correlation does not mean
# causation, it can be inferred that teams that had higher field goal percentages, also had higher free throw percentages.
#
# This correlation reflects how good the Minnesota Timberwolves were as shooting team overall. Meaning, they were successful
# at making shots from anywhere on the court, whether it be from the field or from the free throw line.
```

```python
# Create our x and y values that will be displayed on the graph the FG% and FT% and store them into their respective
# variables 'x' and 'y'
x = timberwolvesFilteredTeamStatsDF['FG%']
y = timberwolvesFilteredTeamStatsDF['FT%']
seasons = timberwolvesFilteredTeamStatsDF['Season']

# This block of code creates the actual scatterplot that we want to create from the above correlation matrix.
# Used bits and pieces data viz notebook on Canvas for help with all visualizations in this notebook.
plt.scatter(x, y)
plt.title('Minnesota Timberwolves FG% and FT% from 2014-2019')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Free Throw Percentage (FT%)')
# The following loop gets us the data labels of each year to appear on their respective data points. .iloc indexing used
# to get the positional index of the season years.
for (i, season) in enumerate(seasons):
    plt.annotate(str(season), (x.iloc[i], y.iloc[i]))

# This block of code creates the fitted regression line so that we can see the actual correlation.
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='blue')

plt.show()
```

```python
# This data frame will only consist of team stats from 2010-2024 of the NBA. First, we need to make a deep copy.
teamStatsDF3 = teamStatsDF.copy(deep=True)

display(teamStatsDF3)
```

```python
# Now, we want to trim down this DF to show only team stats from 2010-2024.
# This block of code gets the rows 465:1845 (all data from before 2010) and it sets it to the variable rowsToDrop.
rowsToDrop = range(465, 1845)

# Now, we drop those rows with this line of code.
teamStatsDF3 = teamStatsDF3.drop(rowsToDrop, axis =0)

display(teamStatsDF3)
```

```python
# Now, we want to get rid of the rows that are the "League Average".
rowsLeagueAverage = teamStatsDF3['Team'] == 'League Average'

# The '~' inverts boolean logic. So, in this case, it searches the 'Team' column for where it equals 'League Average' and
# it returns a boolean 'True'. Then, the '~' inverts this series. True becomes False and False becomes True. So, it
# effectively selects all rows where the team's name is not "League Average." This inherently returns a DF minus the rows
# equal to 'League Average' in the 'Team' column.
teamStatsDF3 = teamStatsDF3[~rowsLeagueAverage]

# Note how we drop down to 450 total rows. This provides a check to our work from the code above.
display(teamStatsDF3)
```

```python
# Now that we have the rows with the "League Average" out, we should reset the index labels since they are now
# not in sequential order. Added drop=True so the 'old' index labels were not added into the DF as a new column.
teamStatsDF3 = teamStatsDF3.reset_index(drop=True)

display(teamStatsDF3)
```

```python
# Now, we want to create these teams from 2010-2024 and split them up by conference.
# ChatGPT was used here to save time it would have taken to manually type in each of these teams.
westernConferenceTeams = [
    'Oklahoma City Thunder', 'Denver Nuggets', 'Minnesota Timberwolves',
    'Los Angeles Clippers', 'Dallas Mavericks', 'Phoenix Suns',
    'New Orleans Pelicans', 'Los Angeles Lakers', 'Sacramento Kings',
    'Golden State Warriors', 'Houston Rockets', 'Utah Jazz',
    'Memphis Grizzlies', 'San Antonio Spurs', 'Portland Trail Blazers']

easternConferenceTeams = [
    'Atlanta Hawks', 'Boston Celtics', 'Brooklyn Nets', 'Chicago Bulls',
    'Charlotte Hornets', 'Cleveland Cavaliers', 'Detroit Pistons',
    'Indiana Pacers', 'Miami Heat', 'Milwaukee Bucks', 'New York Knicks',
    'Orlando Magic', 'Philadelphia 76ers', 'Toronto Raptors', 'Washington Wizards']

# Now we create the dictionaries to map these teams to said conference.
# On the "value" side of the dictionary, we created a loop to iterate through the two 'Conference' variables from above.
teamToConference = {i: 'Eastern' for i in easternConferenceTeams}
teamToConference.update({i: 'Western' for i in westernConferenceTeams})

# Now, we need to map this new column onto the end of our dataframe.
teamStatsDF3['Conference'] = teamStatsDF3['Team'].map(teamToConference)

# This new column is added to the very end of the DF
display(teamStatsDF3)
```

```python
# Some teams from the past decade have switched locations/names like the New Jersey Nets (Brooklyn Nets),
# Charlotte Bobcats (Charlotte Hornets), and New Orleans Hornets (New Orleans Pelicans).
#
# Now, we are going to fill these NA's with their respective conference from that time period.

teamStatsDF3['Conference'] = teamStatsDF3['Conference'].fillna('Eastern')

# This now shows that we have all NA values accounted for including the NA values from the 'Conference'
# column that we just fixed.
teamStatsDF3.info()
```
